```python
# prepare a relative small DataFrame, 
# it is useful to limit the range of statistics only for the year of 2015.

# limit the data for the vote higher than 6.0
movies_2015_vote_low = movies_2015[movies_2015.vote_average>=6] 

# lower than high top voting
movies_2015_vote = movies_2015_vote_low[movies_2015_vote_low.vote_average<=movies['vote_average'].max()+0.1]  
```

```python
# 1, generate the "条形图"
base_color = sb.color_palette()[0]
plt.figure(figsize = [16, 8])
sb.countplot(data = movies_2015_vote, x = 'vote_average', color = base_color)
plt.title('the vote_average count number by the movie year of 2015')

# from bellow bar chart, we can find the most voting value is "6.2", 
# the top voting value within 2015 is "8.9".
```

```python
# 2, generate the "直方图"
base_color = sb.color_palette()[0]
plt.figure(figsize = [16, 8])
bin_edges= np.arange(5.9,movies_2015_vote['vote_average'].max()+0.05,0.05)
plt.hist(data = movies_2015_vote,x = 'vote_average', color = base_color, bins = bin_edges)
plt.title('the vote_average hist count number by the movie year of 2015')
plt.xlabel('vote_average')
plt.ylabel('count')

# from bellow histogram, we can also find the top voting value is lower than "8.9", 
# the most voting rate of votes are croweded nearly the value of "6.2"
```

```python
# 3, generate the "饼图"
sorted_counts = movies_2015_vote['vote_average'].value_counts()
plt.figure(figsize = [30, 15])
plt.pie(sorted_counts, labels = sorted_counts.index,startangle=90, counterclock = False);
plt.axis('square')
plt.title('the vote_average count number by the movie year of 2015')
plt.xlabel('vote_average')

# for the pie chart, the data is placed in the circle with clockwise order by the sorting value.
# the most voting value is "6.2", which is listed in the right top of the pie, 
# the least voting value is listed in the left top of the pie.
```

```python
# generate the box plot of "箱线图"
plt.figure(figsize = [16, 8])
sb.boxplot(data = movies_2015_vote, x = 'vote_average', y = 'vote_count')
plt.title('the box plot for the vote_average count by the movie year of 2015')

# in the bellow boxplot, value of max, min, median, interquartile range shows in the chart.
# some boxes in the left are flat, 
# seems that the data in the DataFrame with some very big values or very small values.
```

```python
# recover the original DataFrame of movies. for the bellow charts.
movies_2015_vote_low = movies_2015[movies_2015.vote_average>=6]
movies_2015_vote = movies_2015_vote_low[movies_2015_vote_low.vote_average<=movies['vote_average'].max()+0.1]
```

```python
# generate the heat map "热力图 " between values:
plt.figure(figsize = [16, 8])
plt.hist2d(data = movies_2015_vote, x = 'vote_average', y = 'vote_count')
plt.colorbar();
plt.title('the vote_average count number by the movie year of 2015')
plt.xlabel('vote average')
plt.ylabel('vote count')

# the heat maps shows that most count number is less than 1000.
# the heat maps also shows that count number is higher from the vote_average=6.0,
# but lower near the number of vote_average=8.5
```

```python
# generate the "轴须图"
g = sb.JointGrid(data = movies_2015_vote, x = 'vote_average', y = 'vote_count',xlim=[5.9,9],ylim=[0,6000])
g.plot_joint(plt.scatter).set_axis_labels("vote_average", "vote_count")
g.plot_marginals(sb.rugplot, height = 0.25)
```

```python
# generate the "带状图"
plt.figure(figsize = [16, 5])
base_color = sb.color_palette()[0]

# left plot: strip plot
plt.subplot(1, 2, 1)
sb.stripplot(data = movies_2015_vote, x = 'vote_average', y = 'vote_count', color = base_color)
plt.title('the vote_average count by the movie year of 2015')
# right plot: swarm plot
plt.subplot(1, 2, 2)
sb.swarmplot(data = movies_2015_vote, x = 'vote_average', y = 'vote_count', color = base_color)
plt.title('the vote_average count by the movie year of 2015')

# the count is higher near the value of vote_average=6.2 and vote_average=7.2
```

```python
# 补充回答问题2：电影票房和知名度的关系
plt.figure(figsize = [16, 5])
plt.scatter(data = movies, x = 'revenue', y = 'popularity')
plt.title('the relation between popularity and the renenue for the movies')
plt.xlim((1000000, movies['revenue'].max()))
plt.ylim((0,movies['popularity'].max()))
plt.xlabel('revenue (USD multiply with 1000000000)')
plt.ylabel('popularity')
```

```python
# generate the "堆积图" 
# we can investigate the voting change from year of 2014 to the year of 2015.
# reference: http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.stack.html
# reference: http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html
movies_2014_2015 = pd.DataFrame(movies, index = list(range(1328)), columns=['vote_average','release_year','vote_count'])
stack = movies_2014_2015.groupby(['vote_average','release_year']).vote_count.sum() 
stack.unstack().plot(kind='bar', stacked=True, color=['darkseagreen', 'darkcyan'],figsize=(14,7))
plt.title('the vote_average count compared from the movie year of 2014 to 2015')

# the x rows is the vote_average, shows the value from 2.4 ~ 9.0
# we can find most rating part is from 6.1 ~ 8.0
```

**对问题1的分析：**
* 根据2015年的电影打分统计，大部分电影分数集中在6分以上8分以下这个范围，可以看出现代电影的评价还是及格并良好的。但电影分数最集中点是6.2分，说明观众也挺苛刻，基本上就是及格。及格以上的数据平缓下降，可以看出优秀电影随着打分上升，越来越稀少。找个优秀电影也不容易。这些都可以从bar plot和histogram图中看出来。螺旋分布的大饼图中可以看出，半饼都是6分的天下。可是，某些打7分的分数中位数远高于6分，数据波动非常大；打6分的部分，异常值较多，也可能是某个电影胡乱打分，或某些电影有非常特殊的优点或缺点。这些在带状图，轴须图和箱线图中都可以看出来。总体来说，热度最集中的电影是6分左右，电影数量大都在1000以下。
* 在对比了2014年到2015年的打分情况，可以看出总体来说，两年的电影分数分布相似，两年中，极少有质量非常差的电影，也就是打分低于4分以下的。大量的电影集中在6.2分左右和7.2分左右。
* 不过，2015年超过8分的打分明显比2014年少，2014年的电影分数高点接近于7分，说明某些2014年的电影质量略好于2015年，但这个数量跳跃性很大，某些2015年良好电影分数也会超过2014年。
* 统计中，可以看出，历史上打分最高的电影是“The Story of Film: An Odyssey”，高达9.2分，进过查询，也是唯一而不是唯二的高分。历史上收入最高的电影是《阿凡达》，收入高达2781505847美元。
* 从票房和知名度看来，很多知名度很高的电影，票房并不是最高的，而票房最高的电影《阿凡达》的知名度只是最高知名度的三分之一左右

```python
#回答问题2：统计分析一般电影放映时长，最多实现的电影放映时长是多少？
plt.figure(figsize = [16, 5])
plt.subplot(1, 2, 1)
bin_edges = np.arange(50,180+1,1)
plt.hist(data = movies, x = 'runtime', bins = bin_edges)
plt.xlim(50,180)
plt.title('the runtime by the movie year of 2015')
plt.xlabel('runtime')
plt.ylabel('count')

# the left figure is the expension figure from runtime of 82 minutes to runtime of 98 minutes.
plt.subplot(1, 2, 2)
bin_edges = np.arange(50,180+1,1)
plt.hist(data = movies, x = 'runtime', bins = bin_edges)
plt.xlim(82,98)
plt.title('the runtime near the 80~90 minutes by the movie year of 2015')
plt.xlabel('runtime')
plt.ylabel('count')

# from the histogram, we can find the most common runtime is 91 minutes,
# from the histogram, most run time is crowded from 90 minutes to 120 minutes.
```

**任务8：** （挑战）请尝试挑选一组特征，进行多变量的可视化。多变量的可视化能够帮助我们揭示数据之间的关系，例如：电影票房和知名度的关系。

```python
movies_recent_years = movies[movies.release_year>=2000]
movies_target_years = movies_recent_years[movies_recent_years.release_year<2016]
bins = np.arange(85,130+1,1)
g = sb.FacetGrid(data = movies_target_years, col = 'release_year',col_wrap = 4);
g.map(plt.hist,'runtime',bins = bins);
plt.xlabel('runtime')
```

以上16组直方图分析出了不同年代，电影放映时长的分布情况。这16个图分别表示了2000年到2015年放映电影的时长变化。

其中，直方图的横坐标是runtime，为电影时长，纵坐标为具有该时长电影的数量。

这里显示了从85分钟电影到130分钟电影的情况（当然我们可以调整显示其它范围的时长）

显然，2012年到2015年的总体电影时长是比2000年到2003年的电影市场要长，我们可以每次看更长时间的电影。

2012年到2015年，电影为95分钟左右的电影大大数量超过了每部120分钟的电影。

而2000年到2003年之间，95分钟左右的电影和120分钟的电影的数量差别不大。各个时长电影数量相差不那么明显。

可以看出，最近几年拍电影，可看的电影时长的伸缩度比较大。

另外，从图中也可以看着，近些年来，总体电影数量明显增多了。而且是从2000年到2015年是逐年增多的迹象。
